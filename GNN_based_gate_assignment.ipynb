```python
import pandas as pd
import numpy as np
import torch
import torch.nn.functional as F  # keep this alias for activation functions
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
import torch.nn as nn
from collections import defaultdict
from datetime import timedelta

# ----------------------------
# Parameters / constants
# ----------------------------
TURNAROUND_REMOTE_MIN = 90  # minutes threshold: >= -> remote stand
REMOTE_STAND_COUNT = 30     # total 30 remote stands
DEVICE = torch.device("cpu")

# ----------------------------
# 1) Filter target flights
# ----------------------------
target_mask = flights_df["Stand"].isna() & flights_df["Status_DEP"].notna() & flights_df["Gate_DEP"].isna()
target_flights_df = flights_df[target_mask].copy().reset_index(drop=True)

# Parse datetime columns
target_flights_df["Scheduled_Time_ARR_dt"] = pd.to_datetime(target_flights_df["Scheduled_Time_ARR"], errors="coerce")
target_flights_df["Scheduled_Time_DEP_dt"] = pd.to_datetime(target_flights_df["Scheduled_Time_DEP"], errors="coerce")

# ----------------------------
# 2) Prepare bridge & remote stands
# ----------------------------
bridge_sets = []
bridge_sets += [f"B{i}" for i in range(1, 37)]
bridge_sets += [f"C{i}" for i in range(1, 19)]
bridge_sets += [f"D{i}" for i in range(1, 32)]
bridge_sets += [f"D{i}" for i in range(41, 58)]
bridge_sets += [f"E{i}" for i in range(12, 25)]
bridge_sets += [f"F{i}" for i in range(1,10)]
bridge_sets += [f"G{i}" for i in range(1,10)]
bridge_sets += [f"H{i}" for i in range(1,8)]
bridge_sets += [f"M{i}" for i in range(1,8)]
BRIDGE_STANDS = sorted(set(bridge_sets))

REMOTE_STANDS = [
    "R001", "R002", "R003", "R004", "R005", "R006", "R007", "R008", "R009", "R010",
    "R011", "R012", "R013", "R014", "R015", "R016", "R017", "R018", "R019", "R020",
    "R021", "R022", "R023", "R024", "R025", "R026", "R027", "R028", "R029", "R030"
]
ALL_STANDS = BRIDGE_STANDS + REMOTE_STANDS

# Map stand -> gate
bridge_gates = {s: s for s in BRIDGE_STANDS}
# Remote stand mapping
remote_gates = {
    "R001": "C21", "R002": "C22", "R003": "C23", "R004": "C24",
    "R005": "D60", "R006": "D62", "R007": "D63", "R008": "D66",
    "R009": "D68", "R010": "D71", "R011": "D73", "R012": "D74",
    "R013": "D77", "R014": "D78", "R015": "D79", "R016": "D81",
    "R017": "D82", "R018": "D83", "R019": "D84", "R020": "D85",
    "R021": "D86", "R022": "D87", "R023": "E2",  "R024": "E3",
    "R025": "E4",  "R026": "E5",  "R027": "E6",  "R028": "E7",
    "R029": "E8",  "R030": "E9"
}

STAND_TO_GATE = {**bridge_gates, **remote_gates}

# ----------------------------
# 3) Node lists and indices
# ----------------------------
flights_list = target_flights_df.to_dict("records")
n_flights = len(flights_list)
stands_list = ALL_STANDS
n_stands = len(stands_list)
gates_list = sorted(set(STAND_TO_GATE.values()))
n_gates = len(gates_list)

# Index maps for nodes
stand_idx_map = {stands_list[i]: n_flights + i for i in range(n_stands)}
gate_idx_map = {gates_list[i]: n_flights + n_stands + i for i in range(n_gates)}

# ----------------------------
# 4) Build node features
# ----------------------------
def safe_turnaround_minutes(arr_dt, dep_dt):
    if pd.isna(arr_dt) or pd.isna(dep_dt):
        return np.nan
    td = dep_dt - arr_dt
    return td.total_seconds() / 60.0

def airline_from_fnumber(fn):
    if not isinstance(fn, str) or len(fn) < 2:
        return "ZZ"
    return fn[:2]

flight_features = []
for rec in flights_list:
    arr_dt = rec.get("Scheduled_Time_ARR_dt")
    dep_dt = rec.get("Scheduled_Time_DEP_dt")
    turnaround = safe_turnaround_minutes(arr_dt, dep_dt)
    if np.isnan(turnaround):
        t_norm = 0.0
    else:
        t_norm = min(turnaround, 360.0) / 360.0
    airline_code = airline_from_fnumber(rec.get("Flight Number_DEP") or "")
    airline_hash = (sum(ord(c) for c in airline_code) % 10) / 10.0
    has_arrival_flag = 0.0 if pd.isna(arr_dt) else 1.0
    flight_features.append([t_norm, airline_hash, has_arrival_flag, 0.0])

stand_features = []
for s in stands_list:
    is_bridge = 1.0 if s in BRIDGE_STANDS else 0.0
    id_hash = (sum(ord(c) for c in s) % 20) / 20.0
    stand_features.append([is_bridge, id_hash, 0.0, 0.0])

gate_features = []
for g in gates_list:
    is_bridge_gate = 1.0 if g in BRIDGE_STANDS else 0.0
    id_hash = (sum(ord(c) for c in g) % 20) / 20.0
    gate_features.append([is_bridge_gate, id_hash, 0.0, 0.0])

x_np = np.vstack([flight_features, stand_features, gate_features])
x = torch.tensor(x_np, dtype=torch.float, device=DEVICE)

# ----------------------------
# 5) Build edges: flight -> stand; stand -> gate
# ----------------------------
edge_src = []
edge_dst = []

for fi, rec in enumerate(flights_list):
    arr_dt = rec.get("Scheduled_Time_ARR_dt")
    dep_dt = rec.get("Scheduled_Time_DEP_dt")
    turnaround = safe_turnaround_minutes(arr_dt, dep_dt)
    if np.isnan(turnaround):
        needs_remote = False
    else:
        needs_remote = turnaround >= TURNAROUND_REMOTE_MIN

    candidate_stands = REMOTE_STANDS if needs_remote else BRIDGE_STANDS
    for s in candidate_stands:
        s_idx = stand_idx_map[s]
        edge_src.append(fi)
        edge_dst.append(s_idx)

# Stand -> Gate (one-way)
for s in stands_list:
    s_idx = stand_idx_map[s]
    g_id = STAND_TO_GATE[s]
    g_idx = gate_idx_map[g_id]
    edge_src.append(s_idx)
    edge_dst.append(g_idx)

if len(edge_src) == 0:
    raise RuntimeError("No edges created; check data and stand lists.")

edge_index = torch.tensor([edge_src, edge_dst], dtype=torch.long, device=DEVICE)

# ----------------------------
# 6) GCN model
# ----------------------------
class SmallGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels=16):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)

    def forward(self, x, edge_index):
        h = self.conv1(x, edge_index)
        h = F.relu(h)           # use F from import torch.nn.functional as F
        h = self.conv2(h, edge_index)
        return h.view(-1)

model = SmallGCN(in_channels=x.shape[1], hidden_channels=16).to(DEVICE)
model.eval()
with torch.no_grad():
    node_scores = model(x, edge_index)  # length = n_flights + n_stands + n_gates

# ----------------------------
# 7) Select best stand for each flight
# ----------------------------
adj_flight_to_stands = defaultdict(list)
src_list = edge_index[0].cpu().numpy()
dst_list = edge_index[1].cpu().numpy()
for s_idx, d_idx in zip(src_list, dst_list):
    if 0 <= s_idx < n_flights and n_flights <= d_idx < n_flights + n_stands:
        adj_flight_to_stands[s_idx].append(int(d_idx))

assignments = {}
for fi in range(n_flights):
    candidate_stand_indices = adj_flight_to_stands.get(fi, [])
    if not candidate_stand_indices:
        assignments[fi] = (None, None, None)
        continue
    best_s_idx = max(candidate_stand_indices, key=lambda si: node_scores[si].item())
    stand_id = stands_list[best_s_idx - n_flights]
    gate_id = STAND_TO_GATE[stand_id]
    stand_score = node_scores[best_s_idx].item()
    assignments[fi] = (stand_id, gate_id, stand_score)

# ----------------------------
# 8) Write back to flights_df
# ----------------------------
out_df = target_flights_df.copy()
assigned_stands = []
assigned_gates = []
assigned_scores = []
for i in range(n_flights):
    s, g, sc = assignments[i]
    assigned_stands.append(s)
    assigned_gates.append(g)
    assigned_scores.append(sc)

out_df["Assigned_Stand"] = assigned_stands
out_df["Assigned_Gate_DEP"] = assigned_gates
out_df["Assigned_Score"] = assigned_scores

for _, row in out_df.iterrows():
    flight_no = row["Flight Number_DEP"]
    sched_dep = row["Scheduled_Time_DEP"]
    assigned_gate = row["Assigned_Gate_DEP"]
    assigned_stand = row["Assigned_Stand"]
    mask = (
        flights_df["Flight Number_DEP"] == flight_no
    ) & (
        flights_df["Scheduled_Time_DEP"].astype(str) == str(sched_dep)
    ) & (flights_df["Stand"].isna()) & (flights_df["Status_DEP"].notna())
    if mask.any():
        idx0 = flights_df[mask].index[0]
        flights_df.at[idx0, "Stand"] = assigned_stand
        flights_df.at[idx0, "Gate_DEP"] = assigned_gate

print("Assigned gates/stands for", n_flights, "target flights (simulation, model untrained).")
print(out_df[["Flight Number_DEP", "Scheduled_Time_ARR", "Scheduled_Time_DEP", "Assigned_Stand", "Assigned_Gate_DEP", "Assigned_Score"]].head(10))
```

```python
import pandas as pd
import numpy as np

# -------------------------------
# 1. Filter July 1 - July 9 flights
# -------------------------------

# --- Step 1: Ensure the datetime column is in the correct format ---
# Convert 'Scheduled_parsed_local' to pandas datetime objects.
# 'errors="coerce"' will convert invalid strings to NaT.
large_df["Scheduled_parsed_local"] = pd.to_datetime(
    large_df["Scheduled_parsed_local"], errors="coerce", utc=True
)

# --- Step 2: Define the date range for training data ---
# Dates are set as UTC to match the converted 'Scheduled_parsed_local'.
start_date = pd.Timestamp("2025-07-01 00:00:00", tz="UTC")
end_date   = pd.Timestamp("2025-07-09 23:59:59", tz="UTC")

# --- Step 3: Filter the DataFrame by date range ---
# Select rows where 'Scheduled_parsed_local' is within the target range.
df_train = large_df[
    (large_df["Scheduled_parsed_local"] >= start_date) &
    (large_df["Scheduled_parsed_local"] <= end_date)
].copy()

# --- Step 4: Output diagnostics ---
print(f"[Info] Selected {len(df_train)} rows for training from {start_date} to {end_date}")
print(df_train.head(3))  # Preview first 3 rows
print(f"Date range: {df_train['Scheduled_parsed_local'].min()} → {df_train['Scheduled_parsed_local'].max()}")

# -------------------------------
# 2. Diagnostics - check required columns and missing values
# -------------------------------
required_cols = ["Flight Number", "Status", "Scheduled Time", "Gate", "Stand", "Terminal"]
missing_cols = [c for c in required_cols if c not in df_train.columns]
if missing_cols:
    raise ValueError(f"Missing required columns: {missing_cols}")

print("[Diagnostics] Missing values per column:")
print(df_train[required_cols].isnull().sum())
```

```python
# ------------------------------
# GNN Train: predict stand for flights (per-flight candidate classification)
# ------------------------------
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from collections import defaultdict
import numpy as np

# Hyperparameters
DEVICE = torch.device("cpu")
TURNAROUND_REMOTE_MIN = 90.0                       # minutes threshold to prefer remote
EPOCHS = 30
LR = 1e-3
HIDDEN = 32

# ------------------------------
# 0) Safety checks
# ------------------------------
# Use only flights that have a known Stand (we need a label)
df_labelled = df_train[df_train["Stand"].notna()].copy().reset_index(drop=True)
if len(df_labelled) == 0:
    raise ValueError("No labelled flights found in df_train (no Stand values).")

print(f"[Info] Labelled training flights: {len(df_labelled)} (out of {len(df_train)} total rows)")

# ------------------------------
# 1) Compose node lists (flights | stands | gates)
# ------------------------------
# flights -> use df_labelled to build training graph
flights_list = df_labelled.to_dict("records")
n_flights = len(flights_list)

# stands and gates use the configuration (BRIDGE_STANDS, REMOTE_STANDS, STAND_TO_GATE)
stands_list = list(BRIDGE_STANDS) + list(REMOTE_STANDS)
n_stands = len(stands_list)
gates_list = sorted(set(STAND_TO_GATE[s] for s in stands_list))
n_gates = len(gates_list)

# Index mapping in concatenated node vector: [ flights | stands | gates ]
idx_flight_base = 0
idx_stand_base = n_flights
idx_gate_base  = n_flights + n_stands

stand_idx_map = {stands_list[i]: idx_stand_base + i for i in range(n_stands)}
gate_idx_map  = {gates_list[i]:  idx_gate_base  + i for i in range(n_gates)}

print(f"[Info] n_flights={n_flights}, n_stands={n_stands}, n_gates={n_gates}")

# ------------------------------
# 2) Build node features (simple numeric features)
# ------------------------------
# safe_turnaround_minutes and airline_from_fnumber are already defined in cell RcTNIwuDFqnf

flight_features = []
for i, rec in enumerate(flights_list):
    arr_dt = rec.get("Scheduled_Time_ARR_dt") if "Scheduled_Time_ARR_dt" in rec else pd.to_datetime(rec.get("Scheduled Time"), errors="coerce")
    dep_dt = rec.get("Scheduled_Time_DEP_dt") if "Scheduled_Time_DEP_dt" in rec else pd.to_datetime(rec.get("Scheduled Time"), errors="coerce")
    turn_min = safe_turnaround_minutes(arr_dt, dep_dt)

    # normalized turnaround (cap at 360 min)
    if np.isnan(turn_min):
        t_norm = 0.0
    else:
        t_norm = min(turn_min, 360.0) / 360.0

    airline_code = airline_from_fnumber(rec.get("Flight Number") or "")
    airline_hash = (sum(ord(c) for c in airline_code) % 10) / 10.0
    has_arrival_flag = 0.0 if pd.isna(arr_dt) else 1.0
    flight_features.append([t_norm, airline_hash, has_arrival_flag, 0.0])

stand_features = []
for s in stands_list:
    is_bridge = 1.0 if s in BRIDGE_STANDS else 0.0
    id_hash = (sum(ord(c) for c in s) % 20) / 20.0
    stand_features.append([is_bridge, id_hash, 0.0, 0.0])

gate_features = []
for g in gates_list:
    is_bridge_gate = 1.0 if g in BRIDGE_STANDS else 0.0
    id_hash = (sum(ord(c) for c in g) % 20) / 20.0
    gate_features.append([is_bridge_gate, id_hash, 0.0, 0.0])

# Stack to one feature matrix and tensor
x_np = np.vstack([flight_features, stand_features, gate_features])
x = torch.tensor(x_np, dtype=torch.float, device=DEVICE)

# ------------------------------
# 3) Build edges: flight -> stands; stand -> gate (single direction)
# ------------------------------
edge_src = []
edge_dst = []

def add_edge(u, v):
    edge_src.append(u); edge_dst.append(v)

for fi, rec in enumerate(flights_list):
    # Use same candidate logic (prefer remote if long turnaround)
    arr_dt = rec.get("Scheduled_Time_ARR_dt") if "Scheduled_Time_ARR_dt" in rec else pd.to_datetime(rec.get("Scheduled Time"), errors="coerce")
    dep_dt = rec.get("Scheduled_Time_DEP_dt") if "Scheduled_Time_DEP_dt" in rec else pd.to_datetime(rec.get("Scheduled Time"), errors="coerce")
    turn_min = safe_turnaround_minutes(arr_dt, dep_dt)

    if pd.isna(turn_min):
        needs_remote = False
    else:
        needs_remote = turn_min >= TURNAROUND_REMOTE_MIN

    candidate_stands = REMOTE_STANDS if needs_remote else BRIDGE_STANDS
    # can also allow both: candidate_stands = BRIDGE_STANDS + REMOTE_STANDS
    for s in candidate_stands:
        s_idx = stand_idx_map[s]
        add_edge(fi, s_idx)

# Stand -> Gate (single direction)
for s in stands_list:
    s_idx = stand_idx_map[s]
    g_id = STAND_TO_GATE[s]
    g_idx = gate_idx_map[g_id]
    add_edge(s_idx, g_idx)

# Create edge_index
edge_index = torch.tensor([edge_src, edge_dst], dtype=torch.long, device=DEVICE)
if edge_index.numel() == 0:
    raise RuntimeError("No edges were created. Check your lists.")

# ------------------------------
# 4) Prepare training labels and candidate lists
# ------------------------------
# For each labelled flight, get the ground-truth stand node index (if mapping exists)
train_flight_nodes = []            # node index of flight (0..n_flights-1)
train_true_stand_node = []         # stand node index (absolute in graph)
adj_candidates = {}                # map flight_node -> list of candidate stand node indices

# Build adjacency lists flight -> candidates from edge_index
src_np = edge_index[0].cpu().numpy()
dst_np = edge_index[1].cpu().numpy()
for u, v in zip(src_np, dst_np):
    if (0 <= u < n_flights) and (idx_stand_base <= v < idx_gate_base):
        adj_candidates.setdefault(int(u), []).append(int(v))

# For each flight (row in df_labelled), if its recorded Stand is in stand list, keep it as training example
for fi, rec in enumerate(flights_list):
    true_stand = rec.get("Stand")
    if pd.isna(true_stand):
        continue
    true_stand = str(true_stand).strip().upper()
    if true_stand not in stand_idx_map:
        continue
    flight_node = fi
    stand_node = stand_idx_map[true_stand]

    candidate_nodes = adj_candidates.get(flight_node, [])
    if not candidate_nodes:
        # no candidate stands for this flight (skip)
        continue
    if stand_node not in candidate_nodes:
        # Append true stand to candidate list so model can learn it
        candidate_nodes = candidate_nodes + [stand_node]
        adj_candidates[flight_node] = candidate_nodes

    train_flight_nodes.append(flight_node)
    train_true_stand_node.append(stand_node)

print(f"[Info] Training examples prepared: {len(train_flight_nodes)}")

if len(train_flight_nodes) == 0:
    raise RuntimeError("No training examples (no labelled flights matched to candidate stands).")

# ------------------------------
# 5) Model definition (same SmallGCN)
# ------------------------------
class SmallGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels=HIDDEN):
        super().__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, 1)

    def forward(self, x, edge_index):
        h = self.conv1(x, edge_index)
        h = F.relu(h)
        h = self.conv2(h, edge_index)
        return h.view(-1)   # scalar score per node

model = SmallGCN(in_channels=x.shape[1]).to(DEVICE)
opt = torch.optim.Adam(model.parameters(), lr=LR)
loss_fn = nn.CrossEntropyLoss()   # will apply it per-flight over candidate logits

# ------------------------------
# 6) Training loop (per-flight cross-entropy)
# ------------------------------
model.train()
for epoch in range(EPOCHS):
    opt.zero_grad()
    scores = model(x, edge_index)    # node-level scores, shape [num_nodes]

    losses = []
    acc_count = 0
    for fi, true_st_node in zip(train_flight_nodes, train_true_stand_node):
        candidates = adj_candidates[fi]
        # indices of candidates in the graph
        cand_tensor = torch.tensor(candidates, dtype=torch.long, device=DEVICE)
        # logits for candidate stand nodes
        cand_logits = scores[cand_tensor]            # shape [num_candidates]
        # target: index position of true_st_node within candidates
        try:
            target_idx = candidates.index(true_st_node)
        except ValueError:
            # safety: skip this example if mismatch
            continue
        loss = loss_fn(cand_logits.unsqueeze(0), torch.tensor([target_idx], device=DEVICE))
        losses.append(loss)

        # accuracy counting (on current weights)
        with torch.no_grad():
            pred_idx = int(torch.argmax(cand_logits).item())
            if candidates[pred_idx] == true_st_node:
                acc_count += 1

    if len(losses) == 0:
        raise RuntimeError("No per-flight losses collected; check candidate lists.")

    batch_loss = torch.stack(losses).mean()
    batch_loss.backward()
    opt.step()

    if (epoch + 1) % 5 == 0 or epoch == 0:
        train_acc = acc_count / len(train_flight_nodes)
        print(f"Epoch {epoch+1:03d}/{EPOCHS}  loss={batch_loss.item():.4f}  acc={train_acc:.3f}")

# ------------------------------
# 7) Save model & do a simple inference pass (assign for training flights)
# ------------------------------
torch.save(model.state_dict(), "smallgcn_gate_assigner.pth")
print("[Info] model saved -> smallgcn_gate_assigner.pth")

model.eval()
with torch.no_grad():
    final_scores = model(x, edge_index)

# Build assignments for training flights and show a sample
train_assigns = []
for fi in range(n_flights):
    cands = adj_candidates.get(fi, [])
    if not cands:
        train_assigns.append((None, None, None))
        continue
    c_tensor = torch.tensor(cands, dtype=torch.long, device=DEVICE)
    logits = final_scores[c_tensor]
    best_pos = int(torch.argmax(logits).item())
    best_node = cands[best_pos]
    stand_id = stands_list[best_node - idx_stand_base]
    gate_id = STAND_TO_GATE[stand_id]
    train_assigns.append((stand_id, gate_id, float(logits[best_pos].item())))

# show first 10
for i in range(min(10, n_flights)):
    print(f"Flight {i} -> Stand/Gate assigned (train) : {train_assigns[i]}")
```

```python
# Compare training assignments with original data

# 1. Create a DataFrame from train_assigns for easier merging
train_assigns_df = pd.DataFrame(train_assigns, columns=["Pred_Stand", "Pred_Gate_DEP", "Pred_Score"])

# 2. Add the predicted assignments back to the df_labelled DataFrame
df_compare = df_labelled.copy()
df_compare = pd.concat([df_compare, train_assigns_df], axis=1)

# 3. Extract relevant columns from raw_df for comparison
raw_df_subset = raw_df[["Flight Number_DEP", "Scheduled_Time_DEP", "Stand", "Gate_DEP"]].copy()
raw_df_subset.rename(columns={"Flight Number_DEP": "Flight Number",
                              "Scheduled_Time_DEP": "Scheduled Time",
                              "Stand": "True_Stand",
                              "Gate_DEP": "True_Gate_DEP"}, inplace=True)

# Ensure the time columns are in a comparable string format
df_compare["Scheduled Time_str"] = df_compare["Scheduled Time"].astype(str)
raw_df_subset["Scheduled Time_str"] = raw_df_subset["Scheduled Time"].astype(str)

comparison_df = pd.merge(
    df_compare,
    raw_df_subset,
    on=["Flight Number", "Scheduled Time_str"],
    how="left", # Use left merge to keep all flights from df_compare (training set)
    suffixes=("_pred", "_true")
)

# Drop the temporary string column used for merging
comparison_df.drop(columns=["Scheduled Time_str"], inplace=True)

# 4. Compare predicted and true values
comparison_df["Stand_Correct"] = comparison_df["Pred_Stand"].astype(str) == comparison_df["True_Stand"].astype(str)
comparison_df["Gate_Correct"] = comparison_df["Pred_Gate_DEP"].astype(str) == comparison_df["True_Gate_DEP"].astype(str)

# Display results
print("Comparison of Predicted vs True Stand/Gate for Training Flights:")
display(comparison_df[["Flight Number", "Scheduled Time_true", "True_Stand", "Pred_Stand", "Stand_Correct", "True_Gate_DEP", "Pred_Gate_DEP", "Gate_Correct"]].head())

# Also show overall accuracy for Stands and Gates, only for non-null pairs
valid_stand_comparisons = comparison_df.dropna(subset=["Pred_Stand", "True_Stand"])
stand_accuracy = valid_stand_comparisons["Stand_Correct"].mean() if not valid_stand_comparisons.empty else 0

valid_gate_comparisons = comparison_df.dropna(subset=["Pred_Gate_DEP", "True_Gate_DEP"])
gate_accuracy = valid_gate_comparisons["Gate_Correct"].mean() if not valid_gate_comparisons.empty else 0

print(f"\nOverall Stand Prediction Accuracy (Training Data, non-null pairs): {stand_accuracy:.4f}")
print(f"Overall Gate Prediction Accuracy (Training Data, non-null pairs): {gate_accuracy:.4f}")
```
